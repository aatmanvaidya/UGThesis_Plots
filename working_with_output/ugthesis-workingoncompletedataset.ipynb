```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("tweets.csv")
```

```python
df.shape
```

```python
df.head()
```

```python
df['NewDateFormat'] = pd.to_datetime(df['tweet_creation'], unit='s')
```

```python
df = df[(df['tweet_creation']>1483228799) & (df['tweet_creation']<=1514764799)]
```

```python
print(df['NewDateFormat'].min())
print(df['NewDateFormat'].max())
```

```python
len(df['user_id'].unique())
```

```python
 #column of all the tweet id's
tweet_id = df['tweet_id']
```

```python
#column of the tweets id's original tweet's of the retweets
rt_id = df['rt_status_id'].dropna().astype(np.int64) 
```

```python
print(len(tweet_id))
print(len(rt_id))
```

```python
tweet_set = set(tweet_id)
rt_set = set(rt_id)
```

```python
print(len(tweet_set))
print(len(rt_set))
```

```python
inter_sets = tweet_set.intersection(rt_set)
```

```python
print(len(inter_sets))
```

```python
intersect = rt_id.isin(tweet_id)
```

```python
intersect.value_counts()
```
