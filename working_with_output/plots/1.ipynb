```python
# Parse the dates in the NewDateFormat column and extract the week number
df["week"] = df["NewDateFormat"].apply(lambda x: datetime.strptime(x, "%Y-%m-%d %H:%M:%S").strftime("%U"))
```

```python
df.head()
```

```python
#lets check month wise distribution of tweets in 2017
print("tweet counts for each month\n")
# Jan
jan_count = df[(df['tweet_creation']>=1483228800) & (df['tweet_creation']<=1485907199)]
print("Jan - " + str(len(jan_count)))
#Feb
feb_count = df[(df['tweet_creation']>=1485907200) & (df['tweet_creation']<=1488326399)]
print("Feb - " + str(len(feb_count)))
#March
mar_count = df[(df['tweet_creation']>=1488326400) & (df['tweet_creation']<=1491004799)]
print("Mar - " + str(len(mar_count)))
# April
apr_count = df[(df['tweet_creation']>=1491004800) & (df['tweet_creation']<=1493596799)]
print("Apr - " + str(len(apr_count)))
# May
may_count = df[(df['tweet_creation']>=1493596800) & (df['tweet_creation']<=1496275199)]
print("May - " + str(len(may_count)))
# June
jun_count = df[(df['tweet_creation']>=1496275200) & (df['tweet_creation']<=1498867199)]
print("Jun - " + str(len(jun_count)))
# July
jul_count = df[(df['tweet_creation']>=1498867200) & (df['tweet_creation']<=1501545599)]
print("Jul - " + str(len(jul_count)))
#Aug
aug_count = df[(df['tweet_creation']>=1501545600) & (df['tweet_creation']<=1504223999)]
print("Aug - " + str(len(aug_count)))
#Sep
sep_count = df[(df['tweet_creation']>=1504224000) & (df['tweet_creation']<=1506815999)]
print("Sep - " + str(len(sep_count)))
#Oct
oct_count = df[(df['tweet_creation']>=1506816000) & (df['tweet_creation']<=1509494399)]
print("Oct - " + str(len(oct_count)))
#Nov
nov_count = df[(df['tweet_creation']>=1509494400) & (df['tweet_creation']<=1512086399)]
print("Nov - " + str(len(nov_count)))
#dec
dec_count = df[(df['tweet_creation']>=1512086400) & (df['tweet_creation']<=1514764799)]
print("Dec - " + str(len(dec_count)))
```
